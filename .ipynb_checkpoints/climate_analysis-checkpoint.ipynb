```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
##Create an engine to connect to sqlite database created in previous step
engine = create_engine("sqlite:///hawaii.sqlite", echo=False)
```

```python
## How many tables are there in the database
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Reflect Database into ORM classes
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
# Map the classes from reflection
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
Station = Base.classes.stations
Measurement = Base.classes.measurements
```

```python
#Import for date manipulations
import datetime as dt
```

```python
##choose a 15 day vacation trip
vac_start_date = '2015-09-01' # start date for Sep 1st 2015
vac_end_date = '2015-09-16'
data_start_date = dt.date(2015,9,1) - dt.timedelta(days=365)
```

```python
##create a database session
session = Session(engine)
```

```python
#Now get the precipitation data from the database for one year before your vacation start date up to your start date for Honululu
record = session.query(Station.station).filter(Station.name.like('%HONOLULU%')).first()
(honolulu_station,) = record


precipitation_data = session.query(Measurement.date,Measurement.prcp).filter(Measurement.station == honolulu_station).\
                    filter(Measurement.date >= str(data_start_date)).filter(Measurement.date <= vac_start_date).all()
```

```python
precipitation_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the results into a pandas dataframe. Set the index to the `date`

precipitation_df = pd.DataFrame(precipitation_data, columns=['Date', 'Precipitation'])
precipitation_df.set_index('Date', inplace=True, )

precipitation_df
```

### Plot the results using the DataFrame plot method:

```python
precipitation_df.plot(kind='area',grid=True)
plt.title(f"Precipitation in Honolulu from {str(data_start_date)} to {vac_start_date}")
plt.xticks(rotation=45)
plt.gray()
plt.show()
```

### Use Pandas to print the summary statistics for the precipitation data:

```python
#calculate min, max, avg and total
avg_pcp = round(precipitation_df["Precipitation"].mean(),2)
total_pcp = round(precipitation_df["Precipitation"].sum(),2)
max_pcp = precipitation_df["Precipitation"].max()
min_pcp = precipitation_df["Precipitation"].min()

pcp_summary_df = pd.DataFrame.from_dict({"Avg":[avg_pcp],"Total":[total_pcp],"Max":[max_pcp],"Min":[min_pcp]})
pcp_summary_df
```

### Design a query to find the most active stations.

#### List the stations and observation counts in descending order

#### Which station has the highest number of observations?

```python

record = session.query(Measurement.station, func.count(Measurement.date)).group_by(Measurement.station).\
            order_by(func.count(Measurement.date).desc()).all()
record
```

#### Print the name of the most active station

```python
most_active_station = record[0][0]
most_active_station_name = session.query(Station.name).filter(Station.station == record[0][0]).first()[0]
most_active_station_name
```

### Design a query to retrieve the last 12 months of temperature observation data (tobs).

```python
latest_date = session.query(Measurement.date).\
order_by(Measurement.date.desc()).first()[0]
latest_date
```

```python
tobs_start_date = dt.datetime.strptime(latest_date,"%Y-%m-%d") - dt.timedelta(days=365)
tobs_start_date = tobs_start_date.strftime("%Y-%m-%d")
```

#### Filter by the station with the highest number of observations:

```python
tobs_data = session.query(Measurement.date,Measurement.tobs).\
    filter(sqlalchemy.and_(Measurement.date <= latest_date, Measurement.date >= tobs_start_date) ).\
           filter(Measurement.station == most_active_station).all()
```

```python
tobs_data
```

```python
tobs_df = pd.DataFrame(tobs_data, columns=['Date', 'tobs'])
tobs_df.set_index('Date', inplace=True )
tobs_df
```

#### Plot the results as a histogram with bins=12:

```python
tobs_df.hist(bins=12)
plt.legend(["tobs"],loc='best')
plt.ylabel("Frequency")
plt.show()
```

#### Write a function called calc_temps that will accept a start date and end date in the format %Y-%m-%d and return the minimum, average, and maximum temperatures for that range of dates.

```python
def calc_temps(start_date, end_date):
    min_temp = session.query(func.min(Measurement.tobs)).\
        filter(Measurement.station == honolulu_station).\
                    filter(Measurement.date >= start_date).\
                        filter(Measurement.date <= end_date).all()[0][0]
    max_temp = session.query(func.max(Measurement.tobs)).\
        filter(Measurement.station == honolulu_station).\
                    filter(Measurement.date >= start_date).\
                        filter(Measurement.date <= end_date).all()[0][0]
    avg_temp = round(session.query(func.avg(Measurement.tobs)).\
        filter(Measurement.station == honolulu_station).\
                    filter(Measurement.date >= start_date).\
                        filter(Measurement.date <= end_date).all()[0][0],2)
    return [min_temp,max_temp,avg_temp]

```

### Use the calc_temps function to calculate the min, avg, and max temperatures for your trip using the matching dates from the previous year (i.e. use "2017-01-01" if your trip start date was "2018-01-01")

```python
obs_start_date = (dt.datetime.strptime(vac_start_date,"%Y-%m-%d") - dt.timedelta(days=365)).strftime("%Y-%m-%d")
temp_ranges = calc_temps(obs_start_date,vac_start_date)
```
