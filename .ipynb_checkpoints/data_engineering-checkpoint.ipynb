```python
# Dependencies
import pandas as pd
import numpy as np
import os
```

```python
# Path of the CSV file
hi_stations_csvfile = "Resources\hawaii_stations.csv"
hi_msrmnts_csvfile = "Resources\hawaii_measurements.csv"
```

```python
# Read CSV files into  pandas DataFrames
stations_df = pd.read_csv(hi_stations_csvfile, dtype=object)
msrmnts_df = pd.read_csv(hi_msrmnts_csvfile, dtype=object)
```

```python
# Preview DataFrame - first stations

stations_df
```

```python
# Preview DataFrame - Now measurements
msrmnts_df.head()
```

```python
#since measurments has missing values. Clean it
msrmnts_df = msrmnts_df.dropna(axis=0)
msrmnts_df.head()
```

```python
#reset the index to remove gaps in the index
msrmnts_df = msrmnts_df.reset_index(drop=True)
msrmnts_df.head()
```

```python
# Save the cleaned data to a file called `clean_hawaii_measurements.csv`
clean_msrmnts_csv = "Resources\clean_hawaii_measurements.csv"
msrmnts_df.to_csv(clean_msrmnts_csv, index=False)
```
